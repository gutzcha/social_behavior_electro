```python
# folder_name = 'D://Project- Electro//social_behavior_electro//analysis//affiliative//population analysis results'
sociability = 'affiliative'
# sociability = 'aversive'

folder_name = osp.join('..','..','analysis',sociability,'spike analysis results')
df_dict_list = glob.glob(osp.join(folder_name, 'Population analysis results for Spikes - Mean firing rate*'))
df_dict_list
```

```python
import pandas as pd

def insert_column_names(df):
    # Copy the column names from the first row
    column_names = df.columns.tolist()
    df.loc[-1] = column_names  # adding a row
    df.index = df.index + 1  # shifting index
    df.sort_index(inplace=True) 

    # Rename the columns as "filenames" and "values"
    df.columns = ["filenames", "values"]

    return df
```

```python
df
```

```python
substrings_to_remove = ['Mean','Before','After', 'During', 'Area','Firing']

df_all_list = []
for df_name in df_dict_list:
    df_dict = pd.read_excel(df_name, None)
    
       
    for sheet, df in df_dict.items():
#         if len(df.columns) > 1:
#             continue
        if "All" in sheet:
            continue
            
        if "Uploaded" in sheet:
            continue
        
        df = insert_column_names(df) # fix table
        df_temp = pd.DataFrame()
        area_name = sheet
        for sub_string_to_remove in substrings_to_remove:
            area_name = area_name.replace(sub_string_to_remove,'')
        area_name = str(area_name)

        vals = df['values']
        filenames = df['filenames']
        
        if "Before" in sheet:
            stage = "1_Spikes_Before"
        elif "During" in sheet:
            stage = "2_Spikes_During"
        elif "After" in sheet:
            stage = "3_Spikes_After"
        else:
            print(sheet)
            raise

        
        df_temp['values'] = vals
        df_temp['area'] = area_name
        df_temp['stage'] = stage
        df_temp['filenames'] = filenames
        
        df_all_list.append(df_temp)
df_all = pd.concat(df_all_list)          
```

```python
df_all.set_index('filenames')
```

```python
df = df_all.copy()
# Create a pivot table to unstack 'freq' and 'stage' columns
pivot_df = pd.pivot_table(df, values='values', index=['filenames'], columns=['area','stage'])

# Reset the index to flatten the columns
pivot_df = pivot_df.reset_index()
pivot_df = pivot_df.set_index(('filenames',""))
pivot_df = pivot_df.rename_axis(index = {('filenames',""):'filenames'})
```

```python
pivot_df
```

```python
# Function to calculate the differences and add columns
def add_difference_columns(df):
    areas = df.columns.get_level_values('area').unique()

    for area in areas:

        before_col = (area, '1_Spikes_Before')
        during_col = (area, '2_Spikes_During')
        after_col = (area, '3_Spikes_After')
        
        # difference 
        diff_during_before = df[during_col] - df[before_col]
        diff_after_before = df[after_col] - df[before_col]

        diff_col_name_during_before = (area,'4_Spikes_DiffDuringBefore')
        diff_col_name_after_before =  (area,'5_Spikes_DiffAfterBefore')
        
        df[diff_col_name_during_before] = diff_during_before
        df[diff_col_name_after_before] = diff_after_before
        
        # ratio        
        ratio_col_name_during_before = (area,'6_Spikes_RatioDuringBefore')
        ratio_col_name_after_before =  (area,'7_Spikes_RatioAfterBefore')
        
        ratio_during_before = (diff_during_before)/(df[before_col])
        ratio_after_before = (diff_after_before)/(df[before_col])


        df[ratio_col_name_during_before] = ratio_during_before
        df[ratio_col_name_after_before] = ratio_after_before

    return df

# Call the function to add difference columns
df_with_diff = add_difference_columns(pivot_df)
```

```python
def sort_columns(df):
    # Sort the columns based on the specified rule
    sorted_columns = sorted(df.columns, key=lambda col: (col[0], col[1]))

    # Reindex the DataFrame with the sorted columns
    sorted_df = df.reindex(columns=sorted_columns)

    return sorted_df


df_with_diff_sorted = sort_columns(df_with_diff)
df_with_diff_sorted.columns
```
